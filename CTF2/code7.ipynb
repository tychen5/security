```python
def prettify_numlist(numbers):
    n = 6
    return '[ ' + '  '.join(map(str,numbers[:n])) + ' ... ' + ' '.join(map(str,numbers[-n:])) + ' ]'
```

```python
all_=[]
for i,kk in enumerate(recvs):
    if i>=22:
        k2 = kk.split(str(i+1)+'\n')[1].split('\n')[0].split('Address: ')[1] #只取address值，3 state
        k2 = textwrap.wrap(k2,8)
        all_.extend(k2)
        k3 = kk.split('BTC\n')[1].split('Password: ')[1].split('\n')[0] #5 state
        k3 = textwrap.wrap(k3,8)
        all_.extend(k3)
        if i == 99:
            print(len(all_))
            break
    
```
